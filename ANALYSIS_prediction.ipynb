```python
# loading 2nd may shared dataset til 16th march
df = pd.read_csv('AIOPS_Predictions_Time_Ahead_V1_28.05.19.csv')
```

```python
df.columns
```

```python
actual_col_li = [col for col in df.columns if 'pred' not in col]
actual_col_li
```

```python
predicted_col_li = [col for col in df.columns if col not in actual_col_li]
predicted_col_li
```

```python
# getting cpu utilisation list

a_cpu_col_li  = [col for col in actual_col_li if 'cpu' in col]
p_cpu_col_li  = [col for col in predicted_col_li if 'cpu' in col]
print(a_cpu_col_li)
print(p_cpu_col_li)
```

```python
# plotting cpu utilisation

f,ax = plt.subplots(6,1,figsize=(30,50))
f.patch.set_facecolor('xkcd:mint green')

for i,(a_col,p_col) in enumerate(zip(a_cpu_col_li,p_cpu_col_li)):
    ax[i].plot(df[a_col])
    ax[i].plot(df[p_col])
    ax[i].set_title('style: {!r}'.format(a_col), color='red')
    ax[i].tick_params(axis='x', colors='red')
    ax[i].tick_params(axis='y', colors='red')
    ax[i].xaxis.label.set_color('red')
    
```

```python
# getting memory utilisation list

a_memory_col_li  = [col for col in actual_col_li if 'memory' in col]
p_memory_col_li  = [col for col in predicted_col_li if 'memory' in col]
print(a_memory_col_li)
print(p_memory_col_li)
```

```python
# plotting memory utilisation

for a_col,p_col in zip(a_memory_col_li,p_memory_col_li):
    plt.plot(df[a_col])
    plt.plot(df[p_col])
    plt.title(a_col)
    plt.show()
```

```python
# plotting disk utilisation

for a_col,p_col in zip(a_disk_col_li,p_disk_col_li):
    plt.plot(df[a_col])
    plt.plot(df[p_col])
    plt.title(a_col)
    plt.show()
```
