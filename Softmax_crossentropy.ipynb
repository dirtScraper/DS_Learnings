```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
li = [2,4,-1,-3]
```

```python
plt.plot(li)
```

```python
def softmax(li):
    ex_li = np.exp(li)
    return ex_li/ex_li.sum()
```

```python
li = softmax(li)
li
```

```python
plt.plot(li)
```
