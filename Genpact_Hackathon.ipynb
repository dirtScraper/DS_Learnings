```python
meal_info = pd.read_csv('meal_info.csv')
```

```python
meal_info.head()
```

```python
def concat_df(main_df,ffc_info,meal_info,on_1,on_2):
    ret_df_1 = pd.merge(main_df,ffc_info,on=on_1)
    ret_df_2 = pd.merge(ret_df_1,meal_info,on=on_2)
    
    return ret_df_2
```

```python
df_merge = concat_df(df,ffc_info,meal_info,'center_id','meal_id')
```
